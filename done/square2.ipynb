```python
from numpy import sqrt
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi, sin, cos, sqrt


plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
plt.rc("font", family="serif", size=16)
```

```python
N = 5000
L = 2.
r1 = lambda u, v : -(u**2 + v**2) - sqrt((u**2 - v**2)**2 - 1)
r2 = lambda u, v : -(u**2 + v**2) + sqrt((u**2 - v**2)**2 - 1)

v = [0, .5,  1., 1.5]
fig, axs = plt.subplots(len(v), figsize=(5, 15), sharex=True)
for i, vv in enumerate(v):
    ex = sqrt(vv**2 + 1)
    u = np.linspace(L, ex)

    ax = axs[i]
    ax.plot(u, r1(u, vv), 'b--')
    ax.plot(-u, r1(u, vv), 'b--')
    ax.plot(u, r2(u, vv), 'k' )
    ax.plot(-u, r2(u, vv), 'k' )

    u0 = np.linspace(-ex, ex)
    ax.plot(u0, - (u0 **2 + vv**2), 'green')
    ax.plot( [ex,ex], [0, -10], 'purple')
    ax.plot([-ex, -ex], [0, -10], 'purple')

    ax.set_ylabel('$r$')
    if i == len(v)-1:
        ax.set_xlabel("$u$")
    ax.set_title("$v = %.1f$"%vv)

    ax.set_xlim(-L, L)
    ax.set_ylim(-8.1, 0.1)

plt.show()
fig.savefig('fig/assym/r1.pdf',bbox_inches="tight")
```

```python
N = 5000
L = 3.
r1 = lambda a, v : -(1 + v**2) - sqrt((1 - v**2)**2 - a**2)
r2 = lambda a, v : -(1 + v**2) + sqrt((1 - v**2)**2 - a**2)
r3 = lambda a, v : -(1 + v**2)*np.ones_like(a)

v = [0, .5, 1.-1e-3, 1.5]
fig, axs = plt.subplots(len(v), figsize=(5, 15), sharex=True)
for i, vv in enumerate(v):
    ex = np.abs(1 - vv**2)
    a = np.linspace(0, L, N)
    b = np.linspace(ex, L, N)

    ax = axs[i]
    ax.plot(r1(a, vv), a, 'b--')
    ax.plot(r2(a, vv), a, 'k' )
    ax.plot(r3(b, vv), b, "green")

    ax.plot([0, -5], [+ex, +ex],'purple')
    ax.plot([0, -5], [-ex, -ex], 'purple') 

    ax.set_ylabel('$\\alpha$')
    if i == len(v)-1:
        ax.set_xlabel("$\\alpha$")
    ax.set_title("$v = %.1f$"%vv)

    ax.set_ylim(0, L)
    ax.set_xlim(-5, 0)

plt.show()
fig.savefig('fig/assym/r3.pdf',bbox_inches="tight")
 
```
