```python
from numpy import sqrt
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi, sin, cos, sqrt


plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
plt.rc("font", family="serif", size=16)
```

```python
N = 5000
L = 2.
r1 = lambda u, v : -(u**2 + v**2) - sqrt((u**2 - v**2)**2 - 1)
r2 = lambda u, v : -(u**2 + v**2) + sqrt((u**2 - v**2)**2 - 1)

v = [0, 1.]
fig, axs = plt.subplots(len(v), figsize=(5, 10), sharex=True)
for i, vv in enumerate(v):
    ex = sqrt(vv**2 + 1)
    u = np.linspace(L, ex)

    ax = axs[i]
    ax.plot(u, r1(u, vv), 'b--')
    ax.plot(-u, r1(u, vv), 'b--') 
    ax.plot(u, r2(u, vv), 'k' )
    ax.plot(-u, r2(u, vv), 'k' )

    u0 = np.linspace(-ex, ex)
    ax.plot(u0, - (u0 **2 + vv**2), 'k')
    ax.plot( [ex,ex], [0, -10], 'purple')
    ax.plot([-ex, -ex], [0, -10], 'purple')

    ax.set_ylabel('$r$')
    if i == len(v)-1:
        ax.set_xlabel("$u$")
    ax.set_title("$v = %.1f$"%vv)

    ax.set_xlim(-L, L)
    ax.set_ylim(-8.1, 0.1)

plt.show()
fig.savefig('fig/assym/r1.pdf',bbox_inches="tight")
```
